```python
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.stats import t
```

# Q-1: Pista House selling Hyderabadi Chicken Dum biryani claims that each parcel packet has 500 grams of biryani (also mentioned on the label of packet). You are sceptic of their claims and believe that on average each packet does not contain 500 grams of biryani. How do you prove your claim?




Null Hypothesis,            H0 = 500

Alternate Hypothesis,       H1 != 500


Collecting Samples of size N

N = 10

N = [490, 420, 480, 500, 499, 496, 498, 508, 480, 491]



```python
pop_mean = 500 
sample = [490, 420, 480, 500, 499, 496, 498, 508, 480, 491]
N = len(sample)
sample_mean = sum(sample)/N
print("Sample mean is {}".format(sample_mean))
```

**Calculating t value**


t-Test-Formula.jpg

```python

def t_value(sample_size, sample_mean, pop_mean, sample_std):
 a = sample_mean - pop_mean
 b = sample_std / sample_size**0.5
 return a/b
```

```python
sample_std = np.std(sample)
print(" standard deviation of the sample is {}". format(sample_std))
```

```python
t_val = t_value(N,sample_mean,pop_mean,sample_std)
print(" t static value is {}".format(t_val))
```

**Confidence Interval**


```python
confidence_interval = 0.95
alpha = 1 - confidence_interval
```

**Deciding Test Type**



```python
t_critical = t.ppf(1-alpha/2,df = N-1)
print("t critical value is {}".format(t_critical))
```

```python
if abs(t_val) >= t_critical:
 print("Reject the null hypothesis.")
else:
 print('Fail to Reject the null hypothesis.')
```

**Plotting**

```python
x_min = 400
x_max = 600
df = N-1


mean = pop_mean
std = sample_std


x = np.linspace(x_min, x_max, 100)
y = t.pdf(x, df,mean, std)
plt.figure(figsize = (15,10))


t_critical_left = pop_mean - (t_critical * std)
t_critical_right = pop_mean + (t_critical * std)


x1 = np.linspace(x_min, t_critical_left, 100)
y1 = t.pdf(x1,df, mean, std)
plt.fill_between(x1, y1, color='blue')
plt.scatter(x1,y1)


x2 = np.linspace(t_critical_right, x_max, 100)
y2 = t.pdf(x2,df, mean, std)
plt.fill_between(x2, y2, color='orange')
plt.scatter(x2,y2)
plt.scatter(sample_mean, 0)
plt.annotate("Sample_mean", (sample_mean-20, 0.0007))
plt.axvline(mean, color='green')
plt.annotate("pop_mean", (mean+1, 0.01))
plt.plot(x, y)
plt.show()
```

Alternate Hypothesis,   H1 : X < 100

N = 10

N = [88, 111, 91, 90, 96, 93, 110, 102, 109, 100]
